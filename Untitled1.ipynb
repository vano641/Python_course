```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
pd.get_dummies(data, columns=['whoAmI'])
```

```python
# Мое Решение:
robot = []
human = []
for i in data['whoAmI']:
  if i=='robot':
    robot += ['True']
    human += ['False']
  else:
    robot += ['False']
    human += ['True']
resDF = pd.DataFrame({'human':human,'robot':robot})
resDF.head(20)
```
